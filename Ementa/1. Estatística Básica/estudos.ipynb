# Variáveis aleatórias contínuas e discretas

**Variável aleatória** é nada mais que uma função que mapeia os resultados de um processo estocástico (aleatório) em **valores numéricos**

- **Variável aleatória discreta**: quando o conjunto de valores é finito ou infinito enumerável
    - Exemplos:
        - número de filhos em uma família
        - resultado do lançamento de um dado
        - quantidade de carros que passam por um pedágio em uma hora

- **Variável aleatória continua**: assume **infinitos** valores em um intervalo.
    - Exemplos:
        - altura de uma pessoa
        - duração de ua chamada telefônica
        - peso de um pacote de arroz

# Função densidade de probabilidade e distribuição acumulada

A **função densididade de probabilidade** descreve a distribuição de probabilidade de uma **variável aleatória** contínua. Ela indica como a probabilidade está distribuída ao longo dos valores possíveis.


Propriedades:

1) Não negatividade: $f(x) \geq 0, \forall x$
2) Área total igual a 1 $\int_{-\infty}^{\infty} f(x) \, dx = 1.$
3) A probabilidade em um intervalo é dada por: $P(a \leq X \leq b) = \int_{a}^{b} f(x)dx$

A **distribuição acumulada F(X)** representa a probabilidade de uma variável aleatória $X$ ser **menor ou igual** a um valor $x$:

$$F(x) = P(X \leq x)$$

# Propriedades de distribuições

## Média

**Definição**: Se trata do valor esperado $E(X)$ e é definido pela soma ponderada de todos os valores possíveis pela respectiva probabilidade. 

**Destaque**: Ela corresponde ao centro de massa da distribuição 

- Discreta: $E(X) = \sum_{i} x_i \cdot  P(X=x)$
- Contínua: $E(X) = \int_{-\infty}^{\infty} x \cdot f(x) dx$

## Mediana

**Definição**: É o valor que divide a distribuição em duas partes iguais (50\% abaixo e acima)

**Destaque**: Ela é uma métrica robusta por não ser afetada por valores extremos (ao contrário da média)

- Discreta: Menor valor $m$ tal que $P(X \leq m) \geq 0.5$
- Contínua: Se dá pela solução de F(m) = 0.5

## Moda

**Definição**: A moda é o valor **mais frequente** da distribuição
- Distribuições unimodais: Uma única moda
- Distribuições multimodais: várias modas.

## Variância ($\sigma^2$ ou VaR(X))

**Definição**: Média dos quadrados dos desvios em relação à média.

- Discreta: $VaR(x) = \sum_{i} (x_i - \mu)^2 \cdot P(X=x_i)$
- Contínua: $\int_{-\infty}^{\infty} (x_i - \mu)^2 f(x) dx$

## Desvio padrão ($\sigma$)

**Destaque**: Medida de dispersão na mesma unidade dos dados

$\sigma = \sqrt{VaR(X)}$


## Amplitude Interquartil (IQR)

**Definição**: Diferença entre o **terceiro** e o **primeiro** quartil ($IQR = Q_3 - Q_1$)

**Interpretação**: Intervalo que contém 50% dos dados centrais

## Quartis e Percentis

Dividem os dados em partes iguais.

**Quartis ($Q_1$, $Q_2$, $Q_3$)**:
- $Q_1$: 25\% dos dados estão abaixo
- $Q_2$ (mediana): 50\% dos dados estão abaixo
- $Q_3$: 75\% dos dados estão abaixo

**Percentis**:

Generalização dos quartis (ex.: 10º percentil = valor abaixo do qual estão 10% dos dados).


## Assimetria e Curtose

### Assimetria

Mede o **grau de deformação** de uma distribuição em relação à simetria. **Indica se os dados estão mais concentrados à esquerda ou à direita da média.**

Tipos:
- Assimetria positiva (cauda à direita): Média > Mediana (Exemplo: distribuição de salário, muitos com salário baixo e poucos com salário alto)
- Assimetria negativa (cauda à esquerda): Média < Mediana (Exemplo: notas de uma prova fácil)
- Simétrica (zero): Média = Mediana (Exemplo: distribuição normal)

Cálculo: 
$$\gamma_1 = \frac{E[(X - \mu)^3]}{\sigma^3}$$


### Curtose

Mede o **achatamento** e o peso das caudas da distribuição em comparação com a normal (ex.: presença de outliers).

Tipo:

- Leptocúrtica ($\gamma_2 > 0$): Caudas pesadas e pico alto (Ex: dados com muitos outliers)
- Mesocúrtica ($\gamma_2 = 0$): Similar à normal
- Platicúrtica ($\gamma_2 < 0$): Caudas leves e achatada (Ex: dados uniformemente distribuídos)

Cálculo:

$$\gamma_2 = \frac{E[(X - \mu)^4]}{\sigma^4} - 3$$

## Código

```python
import numpy as np
from scipy.stats import skew, kurtosis

dados = np.array([6, 7, 7, 9, 10])
print("Média:", np.mean(dados))
print("Mediana:", np.median(dados))
print("Moda:", np.argmax(np.bincount(dados)))
print("Variância:", np.var(dados, ddof=0))     # ddof=0 para população
print("Assimetria:", skew(dados))
print("Curtose:", kurtosis(dados, fisher=True))  # Fisher=True subtrai 3
```

# Principais distribuições 
**Normal, Bernoulli, Binomial, Uniforme, Poisson, Geométrica**

## Normal

- Contínua e simétrica
- Forma de sino
- Definida por média e desvio padrão ($\mu$, $\sigma$)
- É conhecida como **Gaussiana** quando $(\mu=0, \sigma=1)$


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parâmetros
mu, sigma = 0, 1  # média e desvio padrão

# Gerar dados
x = np.linspace(-5, 5, 100)
y = norm.pdf(x, mu, sigma)

# Plotar
plt.figure(figsize=(8, 4))
plt.plot(x, y, 'b-', linewidth=2)
plt.title('Distribuição Normal (μ=0, σ=1)')
plt.xlabel('Valores')
plt.ylabel('Densidade de Probabilidade')
plt.grid(True)
plt.show()
```

## Bernoulli

- Discreta
- Dois resultados possíveis (0 ou 1)
- Definida por p (probabilidade de sucesso)

```python
from scipy.stats import bernoulli

# Parâmetro
p = 0.7  # probabilidade de sucesso

# Gerar dados
x = [0, 1]
y = bernoulli.pmf(x, p)

# Plotar
plt.figure(figsize=(6, 4))
plt.bar(x, y, width=0.1, color=['red', 'blue'])
plt.xticks(x, ['Fracasso (0)', 'Sucesso (1)'])
plt.title('Distribuição de Bernoulli (p=0.7)')
plt.ylabel('Probabilidade')
plt.ylim(0, 1)
plt.grid(axis='y')
plt.show()
```

## Distribuição Binomial

- Discreta
- Número de sucessos em **n** tentativas independentes de Bernoulli
- Definida por $n$ (número de tentativas) e $p$ probabilidade de sucesso

```python
from scipy.stats import binom

# Parâmetros
n, p = 10, 0.5  # tentativas, probabilidade de sucesso

# Gerar dados
x = np.arange(0, n+1)
y = binom.pmf(x, n, p)

# Plotar
plt.figure(figsize=(8, 4))
plt.bar(x, y, color='green', alpha=0.7)
plt.title('Distribuição Binomial (n=10, p=0.5)')
plt.xlabel('Número de Sucessos')
plt.ylabel('Probabilidade')
plt.xticks(x)
plt.grid(axis='y')
plt.show()
```

## Distribuição uniforme

- Contínua ou discreta
- Todos os resultados são equiprováveis
- Definida por $a$ (mínimo) e $b$ (máximo)

```python
from scipy.stats import uniform

# Contínua
a, b = 0, 10  # mínimo e máximo
x = np.linspace(a-1, b+1, 1000)
y = uniform.pdf(x, a, b-a)

plt.figure(figsize=(8, 4))
plt.plot(x, y, 'r-', linewidth=2)
plt.title('Distribuição Uniforme Contínua (a=0, b=10)')
plt.xlabel('Valores')
plt.ylabel('Densidade de Probabilidade')
plt.grid(True)
plt.show()

# Discreta (exemplo de dado)
x_d = np.arange(1, 7)
y_d = [1/6] * 6

plt.figure(figsize=(6, 4))
plt.bar(x_d, y_d, width=0.5, color='purple')
plt.title('Distribuição Uniforme Discreta (Dado Justo)')
plt.xlabel('Face do Dado')
plt.ylabel('Probabilidade')
plt.xticks(x_d)
plt.grid(axis='y')
plt.show()
```

## Distribuição de Poisson

- Discreta
- Número de eventos em um intervalo fixo de tempo/espaço
- Definida por $\lambda$ (taxa média de ocorrência)

```python
from scipy.stats import poisson

# Parâmetro
lambda_ = 4  # taxa média

# Gerar dados
x = np.arange(0, 15)
y = poisson.pmf(x, lambda_)

# Plotar
plt.figure(figsize=(8, 4))
plt.bar(x, y, color='orange', alpha=0.7)
plt.title(f'Distribuição de Poisson (λ={lambda_})')
plt.xlabel('Número de Eventos')
plt.ylabel('Probabilidade')
plt.xticks(x)
plt.grid(axis='y')
plt.show()
```

## Distribuição Geométrica

- Discreta
- Número de tentativas até o primeiro sucesso
- Definida por $p$ (probabilidade de sucesso)

```python
from scipy.stats import geom

# Parâmetro
p = 0.3  # probabilidade de sucesso

# Gerar dados
x = np.arange(1, 15)
y = geom.pmf(x, p)

# Plotar
plt.figure(figsize=(8, 4))
plt.bar(x, y, color='brown', alpha=0.7)
plt.title(f'Distribuição Geométrica (p={p})')
plt.xlabel('Número de Tentativas até o Primeiro Sucesso')
plt.ylabel('Probabilidade')
plt.xticks(x)
plt.grid(axis='y')
plt.show()
```
