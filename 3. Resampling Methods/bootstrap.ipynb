```python
import numpy as np
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

X, y = datasets.load_iris(return_X_y=True)
X.shape, y.shape
```

```python
n_reamostragens = 100
```

```python
model = LogisticRegression()
scores = []

for _ in range(n_reamostragens):
    # Crie a amostra bootstrap (com reposição)
    indices = np.random.choice(range(len(X)), size=len(X), replace=True)
    X_bootstrap, y_bootstrap = X[indices], y[indices]

    # Identifique as amostras que não foram selecionadas (out-of-bag)
    oob_indices = list(set(range(len(X))) - set(indices))
    X_oob, y_oob = X[oob_indices], y[oob_indices]

    # Treine o modelo na amostra bootstrap
    model.fit(X_bootstrap, y_bootstrap)

    # Avalie o modelo no conjunto out-of-bag
    if len(oob_indices) > 0:  # Verifica se há amostras out-of-bag
        y_pred = model.predict(X_oob)
        score = accuracy_score(y_oob, y_pred)
        scores.append(score)

# Resultados
print("Bootstrap Scores:", scores)
print("Mean Score:", np.mean(scores))
print("Standard Deviation:", np.std(scores))
```
